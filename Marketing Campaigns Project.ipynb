```python
# Customers with kids probably have less time to visit a store and would pre
from scipy.stats import pearsonr
import seaborn as sns
%matplotlib inline
SL = 0.05
data = pd.read_csv("C:\\Users\\lokes\\Music\\marketing_data.csv")
df = pd.DataFrame(data)
filtered_data = df[df['Kidhome'] > 0]
web_purchases = filtered_data.groupby('Kidhome')['NumWebPurchases'].sum().reset_index()
store_purchases = filtered_data.groupby('Kidhome')['NumStorePurchases'].sum()
total_web_purchases = filtered_data['NumWebPurchases']
total_store_purchases = filtered_data['NumStorePurchases']
result = pd.merge(web_purchases, store_purchases, on='Kidhome', how='outer')
corr, pvalue = pearsonr(result['NumWebPurchases'] , result['NumStorePurchases'])
if pvalue <= SL:
 print("Alternate Hypothesis (H1) PASSED --- Customers with kids  probably have less time to visit a store and WOULD NOT prefer to shop online")
else:
 print("Null Hypothesis (H0) PASSED -------- Customers with kids probably have less time to visit a store and WOULD NOT prefer to shop online")
```

```python
from scipy.stats import pearsonr
import seaborn as sns
%matplotlib inline
SL = 0.05
channel_columns = data.filter(like='Num')
df = pd.DataFrame(channel_columns)
# Calculate the sum of all columns except 'NumStorePurchases' and add it as 
df['SumExceptStore'] = df.drop(columns=['NumStorePurchases']).sum(axis=1)
# Create a new DataFrame containing only 'SumExceptStore' and 'NumStorePurch
result = df[['SumExceptStore', 'NumStorePurchases']]
corr, pvalue = pearsonr(result['SumExceptStore'] , result['NumStorePurchases'])
if pvalue <= SL:
 print("Alternate Hypothesis (H1) PASSED --- Other distribution channels MAY cannibalize sales at the store")
else:
 print("Null Hypothesis (H0) PASSED -------- Other distribution channels MAY cannibalize sales at the store")
```

```python
import pandas as pd
from scipy.stats import pearsonr
SL = 0.05
data = pd.read_csv("C:\\Users\\lokes\\Music\\marketing_data.csv")
df = pd.DataFrame(data)
df['TotalPurchases'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumStorePurchases']
corr, pvalue = pearsonr((df['Country'] == 'US').astype(int), df['TotalPurchases'])
if pvalue <= SL:
 print("Alternate Hypothesis (H1) PASSED --- The US fares significantly better than the rest of the world in terms of total purchases")
else:
 print("Null Hypothesis (H0) PASSED -------- The US does not fare significantly better than the rest of the world in terms of total purchases")
```

```python
# Which products are performing the best, and which are performing the least
# ============= Wine ============= 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("C:\\Users\\lokes\\Music\\marketing_data.csv")
df = pd.DataFrame(data)
product_columns = data.filter(like='Mnt')
df = pd.DataFrame(product_columns)
# Use the melt function to reshape the data
df = df.melt(var_name='ProductCategory', value_name='TotalRevenue')
# Group by ProductCategory and sum the TotalRevenue
grouped = df.groupby('ProductCategory')['TotalRevenue'].sum().reset_index()
# Sort the DataFrame by TotalRevenue
grouped = grouped.sort_values(by='TotalRevenue', ascending=False)
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(grouped['ProductCategory'], grouped['TotalRevenue'], color='blue')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Revenue by Product Category')
plt.xticks(rotation=45)
# Display the chart
plt.tight_layout()
plt.show()
```
